# Hypothesis

**Experience Level:**  Data scientists with more years of experience will have higher salaries.

**Educational Attainment:** Data scientists with a PhD or Master's degree will command higher salaries than those with a Bachelor's degree.

**Job Title Seniority:** Job titles with "senior," "manager," "lead," or "chief" will be associated with higher salaries compared to "junior" or "analyst" titles.

**Location:** Salaries will be higher in metropolitan areas and regions with a high cost of living.

**Industry Impact:** Data scientists in industries such as finance, technology, and healthcare will have higher salaries than those in other sectors.

**Company Size:** Larger companies will offer higher salaries than smaller ones or startups.

**Skill Specialization:** Proficiency in high-demand skills or tools like AI, machine learning, cloud services, and big data technologies will correlate with higher salaries.

**Remote Work Flexibility:** The option or requirement to work remotely will have an impact on salary, possibly correlating with higher salaries due to broader opportunities.

**Company Revenue:** Companies with higher revenue or funding will tend to offer higher salaries to their data scientists.

**Role Diversity:** Data scientist roles that require a broader set of responsibilities or cross-functional skills will be associated with higher salaries.

_**Note:** **Converting per hour salaries to per year salaries** (assuming 45 work hours per week and 52 working weeks), and removing **employer_provided_salary** from the salary array._

```python
# Removing the string employer provided salary from salary array
salary = salary.apply(lambda x: x if type(x)==type(3.5) else x.lower().replace("employer provided salary:", ""))
salary.unique()[380:]
```

```python
# Converting per hour salaries to per year salaries
def hourly_to_yearly(minlimit, maxlimit):
  x = minlimit.strip()
  y = maxlimit.strip()
  x = int(int(minlimit)*45*52/1000)
  y = int(int(maxlimit)*45*52/1000)
  return '{}-{}'.format(x,y)

salary = salary.apply(lambda x: x if type(x)==type(3.5) else (hourly_to_yearly(x.lower().replace("per hour", "").split('-')[0], x.lower().replace("per hour", "").split('-')[1]) if "per hour" in x.lower() else x))
salary.unique()[380:]
```

```python
# Creating column for salary i.e. average of the range
df['Salary'] = salary.apply(lambda x:  x if type(x)==type(3.5) else (int(x.split("-")[0]) + int(x.split("-")[1].strip()))/2)

plt.figure(figsize=(6,6))
g = sns.distplot(a=df['Salary'])
p = plt.title('Before handling NaN values')
```

**• Feature Selection**<br/>
_Selecting features based on **information gain (mutual_info_regression) and correlation matrix**_

```python
# Splitting the dataset into features and target variable for feature selection
# 'X' contains all features except 'salary', which is our target variable contained in 'y'

X = df.drop('salary', axis=1)
y = df['salary']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression
```

```python
fs = SelectKBest(score_func=mutual_info_regression, k='all')
fs.fit(X, y)
```

```python
def predict_salary(rating, founded, competitors, sector, ownership, job_title, job_in_headquarters, job_seniority, job_skills):
    # Initialize the prediction input list
    prediction_input = []

    # Transform and append the numerical features
    prediction_input.append(sc_rating.transform([[rating]])[0][0])
    prediction_input.append(sc_founded.transform([[founded]])[0][0])
    prediction_input.append(competitors)

    # Encode and append the sector categorical feature
    sector_columns = ['sector_Health Care', 'sector_Business Services',
                      'sector_Information Technology', 'sector_Biotech & Pharmaceuticals']
    sector_values = [1 if 'sector_' + sector == col else 0 for col in sector_columns]
    prediction_input.extend(sector_values)

    # We are not including 'ownership_Private' as per the previous error indicating a mismatch in feature count.
    
    # Encode and append the job title categorical feature
    job_title_columns = ['job_title_data scientist', 'job_title_data analyst']
    job_title_values = [1 if 'job_title_' + job_title == col else 0 for col in job_title_columns]
    prediction_input.extend(job_title_values)

    # Append the job_in_headquarters binary feature
    prediction_input.append(job_in_headquarters)

    # Encode and append the job seniority categorical feature
    job_seniority_map = {'other': 0, 'jr': 1, 'sr': 2}
    prediction_input.append(job_seniority_map[job_seniority])

    # Encode and append the job skills binary features
    skills_columns = ['excel', 'python', 'tableau', 'sql']
    skills_values = [1 if skill in job_skills else 0 for skill in skills_columns]
    prediction_input.extend(skills_values)

    # Ensure the input is in the correct shape for the model
    prediction_input = [prediction_input]  # Convert to 2D array

    # Make the salary prediction using the random forest model
    predicted_salary = rf.predict(prediction_input)[0]

    return predicted_salary

```
